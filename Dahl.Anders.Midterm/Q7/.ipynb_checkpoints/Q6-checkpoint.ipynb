```python
## Adds all the needed python libraries
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from skimage import img_as_ubyte
from PIL import Image

plt.rcParams['image.cmap'] = 'gray' # Display grayscale images in... grayscale.
plt.rcParams['image.interpolation'] = 'none' # Use nearest-neighbour
plt.rcParams['figure.figsize'] = 10, 10 # Default figsize

img = Image.open('images/image1.jpg')
plt.figure()
plt.axis('off')
plt.imshow(img)
colors = img.convert('RGB').getcolors(img.size[0] * img.size[1])
dominant_color = np.array(max(colors,key=lambda item:item[1])[1])

# 0-0-255 blue
# 0-255-0 green
# 255-0-0 red
# 0-255-255 cyan
# 255-0-255 purple
# 255-255-0 yellow
# 255-255-255 white
# 0-0-0 black

displayed_dominant_color = np.array([max(colors,key=lambda item:item[1])[1]])
print displayed_dominant_color
blue = np.array([np.absolute(np.subtract(dominant_color, np.array([0, 0, 257])))])
print blue
green = np.array([np.absolute(np.subtract(dominant_color, [0, 257, 0])) % 256])
red = np.array([np.absolute(np.subtract(dominant_color, [257, 0, 0])) % 256])
cyan = np.array([np.absolute(np.subtract(dominant_color, [0, 257, 257])) % 256])
purple = np.array([np.absolute(np.subtract(dominant_color, [257, 0, 257])) % 256])
yellow = np.array([np.absolute(np.subtract(dominant_color, [257, 257, 0])) % 256])
white = np.array([np.absolute(np.subtract(dominant_color, [257, 257, 257])) % 256])
black = np.array([np.absolute(np.subtract(dominant_color, [0, 0, 0])) % 256])

plt.figure()
plt.axis('off')
plt.imshow(displayed_dominant_color)

imgset1 = np.array([blue, green, red, cyan, purple, yellow, white, black])
plt.figure()
for i,img in enumerate(imgset1):
    plt.subplot(1, len(imgset1), i+1)
    plt.axis('off')
    plt.imshow([img])
```
