# Custom Color Wheel

```python
## Adds all the needed python libraries
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from skimage import img_as_ubyte
from PIL import Image

# Matplot defaults
plt.rcParams['image.cmap'] = 'gray' # Display grayscale images in... grayscale.
plt.rcParams['image.interpolation'] = 'none' # Use nearest-neighbour
plt.rcParams['figure.figsize'] = 5, 5 # Sets the default figure size

# Display the original image and convert it to RGB.
# This should handle other images such as grayscale and hsv.
img = Image.open('images/image1.jpg')
img = img.convert('RGB')
img.thumbnail((128,128), Image.ANTIALIAS)
plt.figure(1)
plt.axis('off')
plt.title('Original image')
plt.imshow(img)
```

## Dominant Color

```python
# Here we find the dominant color in the image - the color that occurs the most often.
colors = img.convert('RGB').getcolors(img.size[0] * img.size[1])
dominant_color = max(colors,key=lambda item:item[1])[1]

# Display the dominant color of the image
displayed_dominant_color = np.array([[dominant_color]])
plt.figure(2)
plt.axis('off')
plt.title('Dominant color from the image')
plt.imshow(displayed_dominant_color)
```

## The Color Wheel - Based off of Boutet's 7-Color Circle from 1708 Applied to the Dominant Color. The color circle is modified slightly.

```python
# 0-0-255 blue
# 0-255-0 green
# 255-255-0 yellow
# 255-128-0 orange
# 255-0-0 red
# 2-74-58 aqua deep
# 257-181-197 violet
# 0, 255, 255 cyan
# 255-255-255 white
# 0-0-0 black

# Calculate the color wheel from the dominant color. This color wheel is based off of
# Boutet's 7-color circle from 1708
blue = np.array([np.absolute(np.subtract(dominant_color, [0, 0, 257])) % 256])
green = np.array([np.absolute(np.subtract(dominant_color, [0, 257, 0])) % 256])
yellow = np.array([np.absolute(np.subtract(dominant_color, [257, 257, 0])) % 256])
orange = np.array([np.absolute(np.subtract(dominant_color, [257, 128, 0])) % 256])
red = np.array([np.absolute(np.subtract(dominant_color, [257, 0, 0])) % 256])
aqua_deep = np.array([np.absolute(np.subtract(dominant_color, [2, 74, 58])) % 256])
violet = np.array([np.absolute(np.subtract(dominant_color, [257, 181, 197])) % 256])
white = np.array([np.absolute(np.subtract(dominant_color, [257, 257, 257])) % 256])
black = np.array([np.absolute(np.subtract(dominant_color, [0, 0, 0])) % 256])

# Display the color wheel so that we can see it.
color_wheel = np.array([blue, green, yellow, orange, red, aqua_deep, violet, white, black])
plt.figure(3, figsize=(20,10))
for i,color in enumerate(color_wheel):
    plt.subplot(1, len(color_wheel), i+1)
    plt.axis('off')
    plt.imshow([color])
```

## Apply Custom Color Wheel to Image

```python
# Find the closest wheel color to the color of the pixel.
def find_closest_wheel_color_fit(image_pixel_color, list_of_wheel_colors):
    list_of_proximites = []
    for c in list_of_wheel_colors:
        diff = np.sum(np.abs(np.subtract(c, image_pixel_color)))
        list_of_proximites.append((diff, c)) 
    return min(list_of_proximites,key=lambda item:item[0])[1][0]
        
# Iterate ALL the pixels and find the closest wheel color. This is slow.
width, height = img.size
img_as_array = np.array(img)
for x in range(height):
    for y in range(width):
        img_as_array[x][y] = np.array(find_closest_wheel_color_fit(img_as_array[x][y], color_wheel))

# Display the image with the custom color wheel colors
plt.figure(4)
plt.axis('off')
plt.title('Image with the Custom Color Wheel Colors')
plt.imshow(img_as_array)
```
